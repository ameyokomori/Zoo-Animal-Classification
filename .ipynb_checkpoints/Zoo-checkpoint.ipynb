```python
from sklearn.model_selection import train_test_split
```

```python
zoo_X = zoo.iloc[:,1:-1]
zoo_y = zoo.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(zoo_X, zoo_y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
# score_linear = cross_val_score(linear, zoo_X, zoo_y, cv=5)
# print(score_linear)
# print(score_linear.mean())
linear.fit(X_train, y_train)
y_pred = linear.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
# scores_knn = cross_val_score(knn, zoo_X, zoo_y, cv=5)
# print(scores_knn)
# print(scores_knn.mean())
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
accuracy_score(y_pred, y_test)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
# score_svc = cross_val_score(svc, zoo_X, zoo_y, cv=5)
# print(score_svc)
# print(score_svc.mean())
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
accuracy_score(y_pred, y_test)
```
